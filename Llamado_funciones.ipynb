```python


# importar las funciones desde r.py
from r import cargar_archivo, rellenar_nulos, ver_nulos, limpiar_atipicos

# cargar csv
df = cargar_archivo("listings1.csv")
print("cargado")

# rellenar nulos
df = rellenar_nulos(df)
print("nulos listos")

# ver nulos
ver_nulos(df)

# limpiar atipicos
df = limpiar_atipicos(df)
print("atipicos fuera")

# resultado
print("final:")
print(df.head())

# guardar
df.to_csv("dataframe_final_procesado.csv", index=False)
print("csv guardado")
```
